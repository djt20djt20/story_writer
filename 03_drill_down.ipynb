```python

from src.chat import Chat
from prompts import *
from tinydb import TinyDB, Query
from src.data_manager import load_data
import json

config = {
    'novel_name': "ai_threat",
    'draft_number': "Draft_1",
    'element_type': 'character',
    'element_name': 'Hero',
    'db_path': "ai_threat.json",
}

prompt_translator = {'character': character}
prompt = prompt_translator[config['element_type']]

db = TinyDB(config['db_path'])
outline = load_data(db, config)
outline_str = json.dumps(outline)
chat = Chat()
chat.set_prompt(prompt)
chat.messages.add_human_message('Summary of novel so far:')
chat.messages.add_human_message(f'We are working on the novel element: {config['element_name']} which is a {config['element_type']}')
chat.messages.add_human_message(outline_str)
chat.continue_conversation()
messages = chat.messages
```

```python
from src.summarizer import ConversationSummarizer
prompt = prompt_translator[config['element_type']]


summarizer = ConversationSummarizer(response_format=character)
messages.add_or_change_system_message(prompt)
summarizer.summarize(messages)
new_summary_str = summarizer.summarized_text
new_summary = summarizer.parse_summary_to_json()
new_summary
```
